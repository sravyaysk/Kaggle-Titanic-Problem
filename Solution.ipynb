# Kaggle Titanic Problem in Python

```python
import pandas as pd
import random
import numpy as np
import csv
import math
```

```python
#read data
dataFrame = pd.read_csv("../KaggleTitanic/train.csv",header = None)
dataFrame
```

```python
data = dataFrame.values
print(data.shape)
```

```python
labels = data[:,1]         #taking survived column
count = 0                  # not survived count

#counting number of survived and not survived from train data
for i in range(labels.shape[0]):
    if(labels[i] == '0'):
        count += 1
        
UnSurvived = count
Survived = labels.shape[0] - count -1 #excluding header
print("total ->",labels.shape[0]-1," Survived ->",Survived,"died ->",UnSurvived)
```

As in our dataset most of the people are dead, let us assume all the people in test data died and check accuracy

```python
# write prediction to csv
with open('../KaggleTitanic/output.csv', 'wt') as f:
    w = csv.writer(f)
    w.writerow(['PassengerId', 'Survived'])
    for i in range(892, 1310):
        w.writerow([i, 0])
```

Now our accuracy is 0.62679

We have 10 features in our dataset. But in titanic disaster all the women and children are given first priority. So lets take Sex and Age variables. 

```python
Gender = data[:,4]   #taking Sex column
count = 0            #Female count

#counting number of male and female from train data
for i in range(Gender.shape[0]):
    if(Gender[i] == 'female'):
        count += 1
        
Female = count
Male = int(Gender.shape[0] - count - 1) #excluding header
print("total ->",Gender.shape[0]-1," Male ->",Male,"Female ->",Female)
```

Now lets see percentage of men and women - alive and dead

```python
#initial count
men_alive = 0
men_dead = 0
women_alive = 0
women_dead = 0
length = data.shape[0]
for i in range(length):
    if(Gender[i] == 'female' and labels[i] == '0'):
        women_dead += 1
    if(Gender[i] == 'female' and labels[i] == '1'):
        women_alive += 1
    if(Gender[i] == 'male' and labels[i] == '0'):
        men_dead += 1
    if(Gender[i] == 'male' and labels[i] == '1'):
        men_alive += 1
print("Women dead",float(women_dead)/Female*100)
print("Women alive",float(women_alive)/Female*100)
print("Men dead", float(men_dead)/Male*100)
print("Men alive", float(men_alive)/Male*100) 
```

That means majority of women alive and majority of men are dead. so in our test set mark all females are survived and all males dead and check the accuracy

```python
#read test data
testdataFrame = pd.read_csv("../KaggleTitanic/test.csv",header = None)
testdataFrame
```

```python
testdata = testdataFrame.values
print(testdata.shape)
testGender = testdata[:,3]
testAge = testdata[:,4]
#print(testGender.shape,testGender[0],testGender[1],testGender[418])
```

```python
# write prediction to csv
with open('../KaggleTitanic/output.csv', 'wt') as f:
    w = csv.writer(f)
    w.writerow(['PassengerId', 'Survived'])
    for i in range(892, 1310):
        if(testGender[i-892] == 'female'):
            w.writerow([i, 1])
        else:
            w.writerow([i, 0])
```

Now our accuracy is 0.7655

```python
# write prediction to csv
with open('../KaggleTitanic/output.csv', 'wt') as f:
    w = csv.writer(f)
    w.writerow(['PassengerId', 'Survived'])
    for i in range(892, 1310):
        if((testGender[i-892] == 'female') and testAge[i-892]<18):
            w.writerow([i, 1])
        else:
            w.writerow([i, 0])
```

```python
#clasification using decision tree
# Importing the required packages
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn import preprocessing

def refine_data(data):
    
    df = pd.read_csv("../KaggleTitanic/train.csv",delimiter=',',header=0)
    # print df.head()
    print df.columns.values
    # print df.dtypes

    #filled empty ages with their median
    df['Age'] = df['Age'].fillna(df["Age"].median())
    df['Fare'] = df['Fare'].fillna(df["Fare"].median())
    # print df.describe()['Age']

    #filled male with 0 and female with 1
    df.loc[df["Sex"]=="male","Sex"] = 0
    df.loc[df["Sex"]=="female","Sex"] = 1
    # print df[df['Sex']==1].count()
    # print df["Sex"]

    # print df["Embarked"].unique()
    # print df.loc[df["Embarked"]=="nan","Embarked"].describe()

    df["Embarked"] = df["Embarked"].fillna("S")
    df.loc[df["Embarked"]=="S","Embarked"]=0
    df.loc[df["Embarked"]=="C","Embarked"]=1
    df.loc[df["Embarked"]=="Q","Embarked"]=2

    #dropped uneccesary variables
    df = df.drop(['PassengerId','Name','Ticket','Cabin',"Embarked"],axis=1)
    # ['PassengerId' 'Pclass' 'Name' 'Sex' 'Age' 'SibSp' 'Parch' 'Ticket' 'Fare','Cabin' 'Embarked']

    print df.head()
    df = np.array(df)
    return df
    
    
# Function importing Dataset
def importdata():
    train_data = pd.read_csv("../KaggleTitanic/train.csv",sep= ',', header = None)
     
    # Printing the dataswet shape
    print ("Dataset Length: ", len(train_data))
    print ("Dataset Shape: ", train_data.shape)
     
    # Printing the dataset obseravtions
    print ("Dataset: ",train_data.head())
    return train_data
 
# Function to split the dataset
def splitdataset(train_data):
    
    train_data = refine_data(train_data)
    X = train_data[:,1:]
    Y = train_data[:,0].tolist()
    
 
    # Spliting the dataset into train and test into 70:30 ratio
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
     
    return X, Y, X_train, X_test, y_train, y_test

# Function to perform training with giniIndex.
def train_using_gini(X_train, X_test, y_train):
 
    # Creating the classifier object
    clf_gini = DecisionTreeClassifier(criterion = "gini",
            random_state = 100,max_depth=3, min_samples_leaf=5)
 
    # Performing training
    clf_gini.fit(X_train, y_train)
    return clf_gini


# Function to perform training with entropy.
def train_using_entropy(X_train, X_test, y_train):
 
    # Decision tree with entropy
    clf_entropy = DecisionTreeClassifier(
            criterion = "entropy", random_state = 100,
            max_depth = 3, min_samples_leaf = 5)
 
    # Performing training
    clf_entropy.fit(X_train, y_train)
    return clf_entropy
 
 
# Function to make predictions
def prediction(X_test, clf_object):
 
    # Predicton on test with giniIndex
    y_pred = clf_object.predict(X_test)
    print("Predicted values:")
    print(y_pred)
    return y_pred
     
# Function to calculate accuracy
def cal_accuracy(y_test, y_pred):
     
    print("Confusion Matrix: ",
        confusion_matrix(y_test, y_pred))
     
    print ("Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
     
    print("Report : ",
    classification_report(y_test, y_pred))
 
# Driver code
def main():
     
    # Building Phase
    data = importdata()
    X, Y, X_train, X_test, y_train, y_test = splitdataset(data)
    clf_gini = train_using_gini(X_train, X_test, y_train)
    clf_entropy = train_using_entropy(X_train, X_test, y_train)
     
    # Operational Phase
    print("Results Using Gini Index:")
     
    # Prediction using gini
    y_pred_gini = prediction(X_test, clf_gini)
    cal_accuracy(y_test, y_pred_gini)
     
    print("Results Using Entropy:")
    # Prediction using entropy
    y_pred_entropy = prediction(X_test, clf_entropy)
    cal_accuracy(y_test, y_pred_entropy)
     
     
# Calling main function
if __name__=="__main__":
    main()
```

Now our accuracy is 80.6 as seen above
